**1. Draw a Line Plot For All The Numerical Columns**


**Pandas uses the plot() method to create diagrams. We can use Pyplot, a submodule of the Matplotlib library to visualize the diagram on the screen**

**data.plot() creates a plot from the DataFrame or Series, using default or specified plot type.**

# plt.show() displays the plot created with Matplotlib.

```python
data.plot() 
plt.show()  
```

**2.Change Size of a Figure & Font: Figure Size : The width of 15 And Height of 10 Font Size:10
**

```python
data.plot(figsize=(15,10), fontsize = 10) 
plt.show()
```

**3. Make Separate Subplots For Each Column**

```python
data.plot(subplots=True, figsize=(15,10), fontsize = 10) #plots each DataFrame column as a separate subplot in a single figure, with a figure size of 15x10 inches and font size of 10 for axis labels and titles.
plt.show()
```

```python
**4. Share X-Axis Values With Other Graphs**
```

```python
#sharex=False: Indicates that the x-axes of the subplots will not be shared, meaning each subplot has its own x-axis
data.plot(subplots=True,figsize=(15,10),fontsize=10,sharex=False)
plt.show()
```

**5. Share X-Axis Values With Other Graphs**

```python
#sharey=True: Indicates that the y-axes of the subplots will be shared, meaning all subplots will use the same y-axis scale.
data.plot(subplots=True,figsize=(15,10),fontsize=10,sharex=False, sharey=True)
plt.show()
```

**6. Draw the Line Plot Only for Petal Width Column**

```python
#data.columns - check the columns for PetalWidthCm
#data[['PetalWidthCm','PetalLengthCm']].plot()  # This is to display 2 columns
data['PetalWidthCm'].plot()
plt.show()
```

**7. Add the Title and Labels to X-axis & Y-axis**

```python
data['PetalWidthCm'].plot(title='Iris Petal Width',xlabel='Indexes', ylabel='Petal width in cm')
plt.show()
```

**8. Draw the Line Plot for Petal Length & Petal Width Columns (Also Change Line Style)**

```python
data[['PetalWidthCm','PetalLengthCm']].plot(style=['--','--'])
plt.show()
```

**9.Draw the Line Plot for Petal Length & Petal Width Columns (Also Change Limit of X-Axis and Y-Axis)**

```python
data[['PetalWidthCm','PetalLengthCm']].plot(style=['--','--'],xlim=(10,140),ylim=(1,15))
```

**10. xticks and yticks**

```python
x = [x for x in range(0,140,10)]
y = [y for y in range(0,10,3)]
print(x,'\n',y)
data[['PetalWidthCm','PetalLengthCm']].plot(style=['--','--'],xticks=x,yticks=y)
plt.show()
```

**11. Area plot**

```python
'''data.plot(kind='area', alpha=0.5) generates an area plot of the DataFrame's data with filled areas under the lines. 
The fill color is semi-transparent with an alpha value of 0.5 (50% transparency)'''

data.plot(kind='area',alpha=0.5)
```

**12. Draw Histogram for SepalLengthCm column**

```python
''' hist() creates a histogram of the DataFrame's data, displaying the distribution of numerical values by dividing them into bins.'''
#data['SepalLengthCm'].plot(kind='hist',bins=20) 
#data['SepalLengthCm'].hist(bins=10) 

data['SepalLengthCm'].plot.hist(bins=20) 
plt.show()
```

**13. Bar Plot**

```python
# This shows for entire dataset
#data.plot.bar() 
# data['Species'].value_counts().plot.bar(rot=60) creates a vertical bar plot of the counts of unique values in the 'Species' column, with

bar_plot = data['Species'].value_counts().plot.bar(rot=60)
for bar in bar_plot.patches:
    bar.set_width(0.15)  # Adjust the width of each bar
plt.show(bar_plot)
```

**14. Scatter plot**

```python
'''It creates a scatter plot to visualize the relationship between 'SepalLengthCm' and 'SepalWidthCm' from the DataFrame, 
with 'SepalLengthCm' on the x-axis and 'SepalWidthCm' on the y-axis'''

data.plot.scatter(x='SepalLengthCm', y='SepalWidthCm')
```

**15. Hexbin plot**

```python
data.columns
```

```python
''' It generates a hexbin plot to visualize the density of points with 'SepalLengthCm' on the x-axis and 'PetalLengthCm' on the y-axis. 
The plot size is set to 10x5 inches'''

data.plot.hexbin(x='SepalLengthCm', y='PetalLengthCm',figsize=(10,5))
```

**16. Box plot**

```python
'''data.plot.box() creates a box plot for each column in the DataFrame, which shows the distribution of the data, 
including quartiles, median, and outliers.'''

data.plot.box()
plt.show()
```

**17. KDE plot**

```python
'''data.plot.kde() creates a Kernel Density Estimate (KDE) plot for each column in the DataFrame, which provides a smooth,
continuous estimate of the data's probability density function'''

data.plot.kde()
plt.show()
```

**18. Density plot**

```python
''' data.plot.density() creates a Kernel Density Estimate (KDE) plot for each column in the DataFrame,
visualizing the smooth probability density function of the data.'''

data.plot.density()
plt.show()
```
